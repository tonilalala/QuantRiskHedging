```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.optimize import minimize
from scipy.optimize import LinearConstraint

Portfolio = pd.read_csv("./Data/portfolio_interpolate1.csv")

#Short List
EFL = pd.read_csv("./Data/EFL.TO.csv")
FCY = pd.read_csv("./Data/FCY.TO.csv")
RXE = pd.read_csv("./DataRXE.TO.csv")
HGU = pd.read_csv("./DataHGU.TO.csv")
HVI = pd.read_csv("./Data/HVI.TO.csv")
XGD = pd.read_csv("./Data/XGD.TO.csv")
ACB = pd.read_csv("./Data/ACB.TO.csv")
CQE = pd.read_csv("./Data/CQE.TO.csv")
IMP = pd.read_csv("./Data/IMP.TO.csv")
WEED = pd.read_csv("./Data/WEED.TO.csv")
XSP = pd.read_csv("./Data/XSP.TO.csv")

AIM = pd.read_csv("./Data/AIM.TO.csv")
BTE = pd.read_csv("./Data/BTE.TO.csv")
TCW = pd.read_csv("./Data/TCW.TO.csv")
UFS = pd.read_csv("./Data/UFS.TO.csv")
CFW = pd.read_csv("./Data/CFW.TO.csv")
YGR = pd.read_csv("./Data/YGR.TO.csv")
CVE = pd.read_csv("./Data/CVE.TO.csv")
DR = pd.read_csv("./Data/DR.TO.csv")
HOU = pd.read_csv("./Data/HOU.TO.csv")
PLI = pd.read_csv("./Data/PLI.TO.csv")
DRT = pd.read_csv("./Data/DRT.TO.csv")
HQU = pd.read_csv("./Data/HQU.TO.csv")
VFF = pd.read_csv("./Data/VFF.TO.csv")

ShortList = [EFL,FCY,RXE,HGU,HVI,XGD,ACB,CQE,IMP,WEED,XSP,AIM,BTE,TCW,UFS,CFW,YGR,CVE,DR,HOU,PLI,DRT,HQU,VFF]
#ShortList = [EFL,FCY,RXE,HGU,HVI,XGD,SEV,ACB,CQE,IMP,MOGO,WEED,XSP]
short_name = ['EFL','FCY','RXE','HGU','HVI','XGD','ACB','CQE','IMP','WEED','XSP','AIM','BTE','TCW','UFS','CFW','YGR','CVE','DR','HOU','PLI','DRT','HQU','VFF']

#Long List
APY = pd.read_csv("./Data/APY.TO.csv")
RTG = pd.read_csv("./Data/RTG.TO.csv")
VLE = pd.read_csv("./Data/VLE.TO.csv")
HUV = pd.read_csv("./Data/HUV.TO.csv")
LongList = [APY,RTG,VLE,HUV]
long_name = ["APY","RTG","VLE","HUV"]
APY_price = APY[['Date','Adj Close']]


tail_days = ['2018-02-02','2018-02-05','2018-02-08','2018-03-22']
tail_index = Portfolio.index.values[Portfolio['date'].apply(lambda x: x in tail_days)]
date = Portfolio['date']


#====================================================
portfolio_price = Portfolio['price']

tail1 = portfolio_price[tail_index[0]]/portfolio_price[tail_index[0]-1]-1
tail3 = portfolio_price[tail_index[2]]/portfolio_price[tail_index[2]-1]-1
tail4 = portfolio_price[tail_index[3]]/portfolio_price[tail_index[3]-1]-1

portfolio_price[tail_index[1]] =portfolio_price[tail_index[2]]
#====================================================
# Normalize data
price_1 = APY['Close']/APY['Close'][0]
price_2 = VLE['Close']/VLE['Close'][0]
# Calculate returns
return_1 = price_1/price_1.shift(1) - 1
return_2 = price_2/price_2.shift(1) - 1
portfolio_return = portfolio_price/portfolio_price.shift(1) - 1
return_1=return_1.drop(tail_index)
return_2=return_2.drop(tail_index)
portfolio_return=portfolio_return.drop(tail_index)

return_1=return_1.dropna()
return_2=return_2.dropna()
portfolio_return=portfolio_return.dropna()

def ABScorr(x):
    abscorr = np.absolute((x[0]*return_1+x[1]*return_2).corr(portfolio_return))
    return(abscorr)
#====================================================
def constaints_tails(price_1,price_2,i,x):
    return (x[0]*(price_1[i]-price_1[i-1])/price_1[i-1]+x[1]*(price_2[i]-price_2[i-1])/price_2[i-1]) + 8*((portfolio_price[i]-portfolio_price[i-1])/portfolio_price[i-1])

x0 = np.array([-0.5,0.5])

def Check_Constraints(price_1,price_2,x):
    check = 0
    for i in [0,1,2,3]:
        Constraint = constaints_tails(price_1,price_2,tail_index[i],x)
        if Constraint < 0:
            #check = check.append(tail_days[i])
            check+=1
    print ('Constraints not satisfied '+str(check))
    return (check)


ResultList = []

for i in range(0,len(ShortList)):
    price_1 = ShortList[i]['Close']/ShortList[i]['Close'][0]
    for j in range(0,len(LongList)):
        price_2 = LongList[j]['Close']/LongList[j]['Close'][0]
        # Calculate returns
        return_1 = price_1/price_1.shift(1) - 1
        return_2 = price_2/price_2.shift(1) - 1
        return_1=return_1.drop(tail_index)
        return_2=return_2.drop(tail_index)

        return_1=return_1.dropna()
        return_2=return_2.dropna()
        #print(return_1.shape)
        #print(return_2.shape)
        #print(portfolio_return.shape)
        def ABScorr(x):
            abscorr = np.absolute((x[0]*return_1+x[1]*return_2).corr(portfolio_return))
            return(abscorr)
        
        cons=({'type': 'ineq',
               'fun': lambda x: constaints_tails(price_1,price_2,tail_index[0],x)},
              {'type': 'ineq',
               'fun': lambda x: constaints_tails(price_1,price_2,tail_index[1],x)},
              {'type': 'ineq',
               'fun': lambda x: constaints_tails(price_1,price_2,tail_index[2],x)},
              {'type': 'ineq',
               'fun': lambda x: constaints_tails(price_1,price_2,tail_index[3],x)},
              {'type': 'ineq',
               'fun': lambda x: 0-x[0]},
              {'type': 'ineq',
               'fun': lambda x: x[0]+1},
              {'type': 'ineq',
               'fun': lambda x: x[1]},
              {'type': 'ineq',
               'fun': lambda x: 1-x[1]})        

        res = minimize(ABScorr, x0, method='COBYLA', constraints=cons)
        check=Check_Constraints(price_1,price_2,res.x)
        if check == 0:
            print("Short ", short_name[i], " long ", long_name[j])
            print(res)
            print(Check_Constraints(price_1,price_2,res.x))
            print(" ")


res.fun

EFL.isnull().sum()
```
